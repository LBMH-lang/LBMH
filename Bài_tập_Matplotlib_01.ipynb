```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x_1 = np.linspace(-6, 5, num=1500)
```

```python
def f(x):
    return x**3 + 3*x + 7
y_1 = f(x_1)
```

```python
fig, ax = plt.subplots()
ax.plot(x_1, y_1)
ax.set_title('y(x)')
plt.legend(['Đường đồ thị'])
```

```python
def f(x):
    return 3*x**2 + 3
y_2 = f(x_1)
```

```python
fig, ax = plt.subplots()
ax.plot(x_1, y_2)
ax.set_title("y'(x)")
plt.legend(['Đường đồ thị'])
```

```python
def f(x):
    return 6*x 
y_3 = f(x_1)
```

```python
fig, ax = plt.subplots()
ax.plot(x_1, y_3)
ax.set_title("y''(x)")
plt.legend(['Đường đồ thị'])
```
