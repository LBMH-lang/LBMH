1 Tải 1 url về

```
from bs4 import BeautifulSoup
import urllib.request
import os
```

```
url= 'https://vietnamnet.vn/'
```

```
page = urllib.request.urlopen(url)
print(page)
```

```
soup = BeautifulSoup(page, 'html.parser')
print(soup)
```
